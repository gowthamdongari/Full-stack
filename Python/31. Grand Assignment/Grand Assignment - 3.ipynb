## Program to check the overlapping of one string's suffix with the prefix of another string 

```python
first_input = str(input())
second_input = str(input())

for char in range(len(first_input)):
    if second_input.startswith(first_input[char:]):
        result = first_input[char:]
        break
    else:
        result = "No overlapping"
print(result)
        
       
```
