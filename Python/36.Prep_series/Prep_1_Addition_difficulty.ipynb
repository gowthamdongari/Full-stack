```python
num_1, num_2 = map(int, input().split(" "))
length_of_num = min(len(str(num_1)), len(str(num_2)))

sum = 0
for i in range(1, length_of_num+1):
    
    sum_of_each_digit = int(str(num_1)[-i]) + int(str(num_2)[-i])
    
    if sum_of_each_digit < 9:
        sum += 1 
    else:
        sum +=2
if sum == length_of_num:
    print("Easy")
else:
    print("Hard")
```
