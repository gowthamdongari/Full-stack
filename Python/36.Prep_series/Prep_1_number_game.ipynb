```python
num_list = list(map(int, input().split(",")))
n = int(input())

num_list_left = num_list.copy()
num_list_right = num_list.copy()


for i in range(n):
    num_list_left.remove(num_list_left[0])
    num_list_right.remove(num_list_right[-1])

left=""
right=""
for i in range(len(num_list_left)):
    left = left + str(num_list_left[i]) + " "
    right = right + str(num_list_right[i]) + " "


print(left)
print(right)
```

```python
num_list = input().split(',')
n = int(input())

num_list_left = num_list[n:]
num_list_right = num_list[:-n]

left=""
right=""
for i in range(len(num_list_left)):
    left = left + num_list_left[i] + " "
    right = right + num_list_right[i] + " "
    
print(left)
print(right)
```
