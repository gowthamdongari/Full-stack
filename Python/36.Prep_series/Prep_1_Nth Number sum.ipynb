```python
n, m = map(int, input().split(' '))
num_list = list(map(int, input().split(' ')))

factor = []
r=m//n
for i in range(1, r+1):
    factor.append(i*n)
sum =0
for i in factor:
    sum = sum + num_list[i-1]
print(sum)
```
