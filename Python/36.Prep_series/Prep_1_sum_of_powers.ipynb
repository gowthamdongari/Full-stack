## Sum of powers

```python
num = "95 99 1005 504"
num_list = list(num.split(' '))
sum = 0
for i in num_list:
    num = i[:-1]
    num = int(num)
    power = i[-1]
    power = int(power)
    result = num**power
    sum = sum + result
print(sum)
    
```

```python
def sum_of_powers():
    num_list = list(input().split(' '))
    sum = 0
    for i in num_list:
        if len(i)>=2:
            num = i[:-1]
            num = int(num)
            power = i[-1]
            power = int(power)
            result = num**power
            sum = sum + result
    return sum
sum_of_powers()
```

```python
def sum_of_powers(intergers_list):
    num_list = intergers_list
    sum = 0
    for i in num_list:
        if len(i)>=2:
            num = i[:-1]
            num = int(num)
            power = i[-1]
            power = int(power)
            result = num**power
            sum = sum + result
    return sum
    
def main():
    intergers_list = input().split()
    powers_sum = sum_of_powers(intergers_list)
    print(powers_sum)

main()
```
