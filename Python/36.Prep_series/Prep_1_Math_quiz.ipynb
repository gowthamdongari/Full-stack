```python
num_list = list(map(int, input().split()))
m = int(input())

for i in range(m):
    range_1, range_2 = map(int, input().split(' '))
    sum = 0
    for num in range(range_1, range_2+1):
        if num in num_list:
            num_count = num_list.count(num)
            sum += num_count*num 
    print(sum)
```
