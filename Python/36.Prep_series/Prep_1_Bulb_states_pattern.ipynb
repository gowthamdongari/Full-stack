```python
def switch_on_off_bulb(n):
    for i in range(1, n+1):
        if i%2 == 1:
            res = ""
            for j in range(1, n+1):
                if  j%2==1:
                    res = res + "1" + " "
                elif j%2==0:
                    res = res + "0" + " "
            print(res)
        elif i%2==0:
            res = ""
            for j in range(1, n+1):
                if  j%2==1:
                    res = res + "0" + " "
                elif j%2==0:
                    res = res + "1" + " "
            print(res)
    
n = int(input())
switch_on_off_bulb(n)
```
